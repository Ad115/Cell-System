```python
t = tree.phylogeny

from ete3 import TreeStyle
ts = TreeStyle()
ts.show_leaf_name = False
ts.mode = "c"
ts.arc_start = -180 # 0 degrees = 3 o'clock
ts.arc_span = 180

t.show(tree_style=ts)
```
